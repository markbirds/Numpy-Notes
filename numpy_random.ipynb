### Binomial Distribution 
[Back to top](#top)

#### Formula for Binary Distribution

### $b(x;n,p) = C(n,x)p^xq^{n-x}$

### $b(x;n,p) = \frac{n!}{x!(n-x)!}p^xq^{n-x}$

b(x;n,p) is the binomial distribution <br>
p is the probability of success <br> 
q = 1-p is the probability of failure <br> 
x is the number of successes desired <br> 
n is the number of trials <br> 
C(n,x) is the binomial coefficient

```python
import math

print('Sample Problem: A biased coin is tossed 6 times. The probability denoted as p on any toss is 0.3.')
print('If we let x denotes the number of heads appear, what is P(x=2)?\n')

print('Given:')
print('p = 0.3')
print('q = 1 – p = 1 – 0.3 = 0.7')
print('x = 2')
print('n = 6')

ans = round(math.factorial(6)/(math.factorial(2)*math.factorial(6-2))*0.3**2*0.7**(6-2),2)
print('\nThe probabilty of 2 heads will appear is ',ans)
```

```python
print('Converting the sample problem above to code: np.random.choice(["H","T"],p=[0.3,0.7],size=6)')

print('\nA biased coin is tossed 6 times: ',np.random.choice(["H","T"],p=[0.3,0.7],size=6))

print('\nTrials')
trials = 0
while True:
    random = np.random.choice(["H","T"],p=[0.3,0.7],size=6)
    print('Trial ',trials,':',random)
    trials+=1
    if len(random[random=="H"]) == 2:
        break
    
print('\nNumber of trials to get 2 heads:',trials)
```

```python
# returns the number of successes
random = np.random.binomial(n=6,p=0.5,size=10)
sns.distplot(random,hist=False)
plt.title('Number of heads appeared out of 6 coins among 10 students')
plt.show()
print(random)
```

### Poisson Distrbution
[Back to top](#top)

#### Formula for Poisson Distribution

### $P(x) = \frac{(e)^{-lam}(lam)^x}{x!}$

P(x) is the poisson distribution <br>
e = 2.71828, a constant used in connection with natural logarithm <br>
lam =  mean or average in the distribution <br>
x is the specific value in which we are interested

```python
print('Example 2. The average number of residence lot sold by the ERIS Company is 2 lots per day.') 
print('What is the probability that exactly 3 lots will be sold tomorrow?\n')

print('lam = 2; since 2 lots are sold per day, on average.') 
print('x = 3; since we want to find the likelihood that 3 lots will be sold tomorrow.') 
print('e = 2.71828; since e is a constant equal to approximately 2.71828.\n')

ans = round((2.71828**-2)*((2)**3)/math.factorial(3),2)
print('The probability of selling 3 lots tomorrow is ',ans)
```

```python
print('Applying numpy.random.poisson:')

print('Number of lots to be sold in the next 10 days:')
x = np.random.poisson(lam=2, size=10)

for index,value in enumerate(x):
    print('Day ',index+1,': ',value)
    
print('\nAverage lots sold for 10 days: ',x.mean())
```

```python
sns.distplot(x)
plt.grid()
plt.show()
```

### Multinomial Distribution
[Back to top](#top)

```python
print('It describes outcomes of multi-nomial scenarios unlike binomial where scenarios must be only one of two.')

x = np.random.multinomial(n=6, pvals=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6])
dice = np.array(['1 dot','2 dots','3 dots','4 dots','5 dots','6 dots'])

print('\nA die was tossed 6 times:\nNumber of occurence of each dot:\n')

for i in range(6):
    print(dice[i],': ',x[i])
```
